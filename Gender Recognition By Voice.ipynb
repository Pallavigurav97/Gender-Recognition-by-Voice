# Machine Learning Algorithms

Here we are going to perform different classifiers like SVM, DT and Logistic Regression algorithm to compare their performances.

### Confusion Matrix:
    A confusion matrix contains information of actual and predicted classification done by a classification model. The performance of such system is commonly evaluated using the data in the matrix.
            [TN  FP]
            [FN  TP]
Where,
True Positive (TP): You predicted positive and it's true.
True Negative (TN): You predicted negative and it's true.
False Positive (FP): You predicted positive and it's false.
False Negative (FN): You predicted positive and it's false.  
    
### Measures of performance evaluation:
#### Accuracy: 
Accuracy is the most intuitive performance measure and it is simply a ratio of correctly predicted observation to the total observations. It is the % of correct classification by the classifier.

            Accuracy = (TP+TN)/(TP+FP+FN+TN)

#### Precision: 
Precision is the ratio of correctly predicted positive observations to the total predicted positive observations.

            Precision = TP/(TP+FP)

#### F1-Score: 
F1 Score is the weighted average of Precision and Recall. Therefore, this score takes both false positives and false negatives into account.

            F1-Score = 2TP/(2TP+FP+FN)

#### Support: 
The support is the number of samples of the true response that lie in that class.


Now, let's determine the dependent (y) and independent (x) variables of the data set:

```python
y=data.label
x_data=data.drop(["label"],axis=1)
x=(x_data-np.min(x_data))/(np.max(x_data)-np.min(x_data))
x = (x_data - np.min(x_data)) / (np.max(x_data)).values
```

Often we need to convert dependent variable values to integer types before creating an ML model. Our dependent variable is the "label" column. And built with string values ("male" and "female"). So we have to change these values. Let's replace "male" ones with 0 and "female" ones with 1 :

```python
data.label=[0 if each=="male" else 1 for each in data.label]
```

## Train Test Split
Before building our ML models, we split the data into training and test data as a final step:

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
##test_size=0.2 means %20 test datas, %80 train datas
```

We are now ready to build machine learning models.

## Logistic Regression:
1] Logistic regression is a statistical method used to analyze a data set with one or more independent variables that define a result.

2] In this method the result is defined by two different values (True-False, Yes-No, 1-0, Cat-Dog, Good-Bad ...).

3] It is used for predicting the probability of a categorical dependent variable using a given set of independent variables.. It is used primarily for predicting binary or multi class dependent variables.

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
print("Test Accuracy: {} %".format(lr.score(x_test,y_test)*100))
```

## Support Vector Machine:
1] It is a supervised learning algorithm based on statistical learning theory.

2] It tries to find the optimal line that separates the two labeled classes.
The optimal line is the furthest away from members of both classes.

3] This model does not have an overfitting problem.

```python
from sklearn.svm import SVC
svm=SVC(random_state=42)
svm.fit(x_train,y_train)
print("Test Accuracy: {} %".format(svm.score(x_test,y_test)*100))
```

## Decision Tree Classification:
1] It is a classification method that creates a model in the form of a tree structure consisting of decision nodes and leaf nodes by feature and target.

2] It can process both numerical and categorical data.

3] It is easy to understand and interpret.

4] This model can address multi-output problems. 

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
print("Test Accuracy: {} %".format(dt.score(x_test,y_test)*100))
```
